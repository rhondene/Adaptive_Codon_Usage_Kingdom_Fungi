# Principal Component Analysis of Relative Synonymous Codon Usage RSCU


```python
#import necessary python libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
#load matrix of transcriptome-wide RSCU for all 459 species
comb_rscu=pd.read_csv('Updated_RSCU_transcriptome-wide(Dec2021).csv')
comb_rscu.head()
```

### Run PCA

```python
#extract codon columns
X = comb_rscu[['Phe-UUU', 'Phe-UUC', 'Ser4-UCU', 'Ser4-UCC', 'Ser4-UCA', 'Ser4-UCG',
       'Ser2-AGU', 'Ser2-AGC', 'Leu4-CUU', 'Leu4-CUC', 'Leu4-CUA', 'Leu4-CUG',
       'Leu2-UUA', 'Leu2-UUG', 'Tyr-UAU', 'Tyr-UAC', 'Cys-UGU', 'Cys-UGC',
       'Arg4-CGU', 'Arg4-CGC', 'Arg4-CGA', 'Arg4-CGG', 'Arg2-AGA', 'Arg2-AGG',
       'Pro-CCU', 'Pro-CCC', 'Pro-CCA', 'Pro-CCG', 'His-CAU', 'His-CAC',
       'Gln-CAA', 'Gln-CAG', 'Ile-AUU', 'Ile-AUC', 'Ile-AUA', 'Thr-ACU',
       'Thr-ACC', 'Thr-ACA', 'Thr-ACG', 'Asn-AAU', 'Asn-AAC', 'Lys-AAA',
       'Lys-AAG', 'Val-GUU', 'Val-GUC', 'Val-GUA', 'Val-GUG', 'Ala-GCU',
       'Ala-GCC', 'Ala-GCA', 'Ala-GCG', 'Asp-GAU', 'Asp-GAC', 'Glu-GAA',
       'Glu-GAG', 'Gly-GGU', 'Gly-GGC', 'Gly-GGA', 'Gly-GGG',]]

#1. perform preprocessing standardization
pca = PCA(n_components=3,random_state=3)
df=X
data_scaled = pd.DataFrame(preprocessing.scale(df),columns = df.columns)
#2. Fit PCA
T = pca.fit_transform(data_scaled)
components_cod= pd.DataFrame(pca.components_,columns=data_scaled.columns,index = ['PC1','PC2','PC3'])
components_cod 
#Convert PCA output to a dataframe table
T_df = pd.DataFrame(T)
T_df.columns = ['PC1','PC2','PC3']
components_cod
#print percent Variance explained
pca.explained_variance_ratio_*100
```
